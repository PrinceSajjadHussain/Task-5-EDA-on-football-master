```python

# Packages to deal with the data
import pandas as pd
import numpy as np
```

```python
# Read data in a dataframe format
dataframe = pd.read_csv('data/英超.csv')
print(dataframe.shape)
dataframe
```

```python
# Convert dataframe to np.array
data = np.array(dataframe)[:,1:]
# Look at the data type on one example
data[0]
```

```python
# Encode some features
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data[:,1] = le.fit_transform(data[:,1])
data[:,3] = le.fit_transform(data[:,3])
data[0]
```

```python
data_X = np.array(data[:,:-1],dtype='float')
data_Y = np.array(data[:,-1],dtype='float')
```

```python
# Split training and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_X, data_Y,
                                                    test_size=0.33, random_state=0)
```

```python
# Model_1: random forest
from sklearn.ensemble import RandomForestRegressor
for max_depth in range(2,10):
    model_1 = RandomForestRegressor(max_depth=max_depth, random_state=0,n_estimators=100)
    model_1.fit(X_train,y_train)
    y_pre = model_1.predict(X_test)
    print(np.mean((y_pre-y_test)**2),'max_depth: ',max_depth)
```

```python
# Model_2: SVM Regression
from sklearn.svm import SVR
for C in [0.1,1.0,10]:
    for kernel in ['rbf','linear']:
        model_2 = SVR(C = C,kernel=kernel,gamma='auto')
        model_2.fit(X_train,y_train)
        y_pre = model_2.predict(X_test)
        print(np.mean((y_pre-y_test)**2),' C=',C,' kernel=',kernel)
```
